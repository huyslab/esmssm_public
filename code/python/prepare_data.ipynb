```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels as sm
from ssm_prep_data import EsmssmDataPrep
prep = EsmssmDataPrep()
```

```python
filepath = '../../data/'
savefiles = False
mood_labels = ['cheerful', 'content', 'anxious', 'sad']
```

```python
file = 'leemput/data_patients.csv'
df_patients = prep.load_data(filepath, file, 'leemput_patients')
file = 'leemput/data_twins.dta'
df_controls = prep.load_data(filepath, file, 'leemput_controls')
```

```python
# Replicate leemput findings

def cat_plot_replication(x, y, var_name, ax, title):
    X = pd.concat((pd.DataFrame(x), pd.DataFrame(y)),axis=1)
    X.columns = ['cheerful', 'anxious', 'content', 'sad', 'group']

    X_melt = pd.melt(X[[['cheerful', 'content'],['anxious', 'sad']][p] + ['group']], id_vars=['group'], \
                     value_vars=[['cheerful', 'content'],['anxious', 'sad']][p], value_name=var_name)
    b = sns.pointplot(data = X_melt, x='group', y=var_name, hue='variable', ax = ax)
    b.set_title(title)

fig, axs = plt.subplots(2,2,figsize=(12,8))

for p, df_specific in enumerate([df_patients, df_controls]):
    ids = np.unique(df_specific['subjno'].dropna())
    AR,SD, G = [], [], []
    for i in ids:
        idx = df_specific['subjno'] == i
        ar = []
        for m in ['opgewkt','angstig','tevreden','somber']:
            tmp = df_specific[m][idx]
            if np.var(tmp.dropna()) != 0 and len(tmp.dropna()) > 2:
                acorr = sm.tsa.stattools.acf(tmp.dropna(), nlags = 1)
                ar.append(acorr[1])
            else:
                ar.append(np.nan)
                    
        AR.append(ar)
        SD.append(df_specific[['opgewkt','angstig','tevreden','somber']][idx].std().values)
        G.append(df_specific[['hdrs_tert', 'dep_mean_tert'][p]][idx].mean(skipna=True))
        
    cat_plot_replication(AR, G, 'autocorrelation (AR1)', axs[0,p], ['patients', 'general population'][p])
    cat_plot_replication(SD, G, 'variance (SD)', axs[1,p], ['patients', 'general population'][p])
    plt.tight_layout();
```

```python
# Data characteristics number of EMA measurments
n_measurements = []
fig, axs = plt.subplots(1,2,figsize=(20,5))
for p, df_specific in enumerate([df_patients, df_controls]):
    ids = np.unique(df_specific['subjno'])
    for i in ids:
        idx = df_specific['subjno'] == i
        tmp = sum((~df_specific[['opgewkt','angstig','tevreden','somber']][idx].isna()).all(axis=1))
        if tmp == 0: 
            continue
        else:
            n_measurements.append(tmp)
    axs[p].hist(n_measurements, bins=20, ec="k")
    axs[p].axvline(x=40, color='red')
    axs[p].set(title=['patients', 'general population'][p], xlabel='number of esm measurements')
```

```python
# Data characteristics depression score distributions
n_measurements = []
fig, axs = plt.subplots(1,2,figsize=(20,5))
for p, df_specific in enumerate([df_patients, df_controls]):
    ids = np.unique(df_specific['subjno'])
    for i in ids:
        idx = df_specific['subjno'] == i
        tmp = df_specific[['hdrs1', 'dep1'][p]][idx].mean(skipna=True)
        if tmp == 0: 
            continue
        else:
            n_measurements.append(tmp)
    axs[p].hist(n_measurements, bins=10, ec="k")
    axs[p].set(title=['patients', 'general population'][p], xlabel='depression score (HDRS)')
```

```python
# Hypothesis mean EMA ratings correlate with depression score
fig,axs=plt.subplots(2,4,figsize=(20,10))
for p, df_specific in enumerate([df_patients, df_controls]):
    ids = np.unique(df_specific['subjno'])
    M,dep = [], []
    for i in ids:
        idx = df_specific['subjno'] == i
        M.append(df_specific[['opgewkt','angstig','tevreden','somber']][idx].mean().values)
        dep.append(df_specific[['hdrs1', 'dep1'][p]][idx].mean(skipna=True))
    X = pd.concat((pd.DataFrame(M), pd.DataFrame(dep)),axis=1)
    X.columns = ['cheerful', 'anxious', 'content', 'sad', 'dep']
    for m, mc in enumerate(X.columns[:-1]):
        b = sns.regplot(data=X, x='dep', y=mc, ax=axs[p,m])
        b.set(title = ['patients', 'general population'][p], xlabel = 'depression score (HDRS)', ylabel = mood_labels[m] + ' mean rating')
        
        
# Suprisingly low in patients.
```

```python
# Prep leemput data
file = 'leemput/data_patients.csv'
df_patients = prep.load_data(filepath, file, 'leemput_patients')
file = 'leemput/data_twins.dta'
df_controls = prep.load_data(filepath, file, 'leemput_controls')

# mood (all columns including mood items)
mood_columns = ['opgewkt', 'tevreden', 'angstig', 'somber']

df_merged, df_exclusion = pd.DataFrame(), pd.DataFrame()
for i, df in enumerate([df_patients, df_controls]):
    df_new, exclusion_code, foo = prep.creating_df(df.copy(), mood_columns, mood_labels, df.name)
    df_new['group'] = i
    df_merged = pd.concat((df_merged, df_new))
    exclusion_code['group'] = i
    df_exclusion = pd.concat((df_exclusion, exclusion_code))
                             
df_merged['id'][df_merged['group']==1] = df_merged['id'][df_merged['group']==1] + \
                             max(df_merged['id'][df_merged['group']==0])

if savefiles:
    df_merged.to_csv(filepath + 'data_prep/leemput/mooddata.csv', \
                     index = False)
    df_exclusion.to_csv(filepath + 'data_prep/leemput/exclusioncode.csv', \
                        index = False)

print('N = ' + str(len(np.unique(df_merged['originalid']))))
```

Preparation of N=1 dataset for Kalman filter
--

```python
df_singledata = pd.read_csv(filepath + 'singledataset/ESMdata.csv')
df_singledata.name = 'singledataset'
df_singledata['subjno'] = 1

# Mood (all columns including mood items)
mood_columns = [col for col in df_singledata.columns if 'mood_' in col]
mood_labels = [item[5:] for item in mood_columns]
idx_different_scaling = df_singledata[mood_columns].min() < 0
list_different_scaling = [mood_columns[i] for i in range(len(mood_columns)) if idx_different_scaling[i]]
df_singledata[list_different_scaling] = df_singledata[list_different_scaling] + 4 
selected_items = [8,3,5,1]
mood_columns = list(mood_columns[i] for i in selected_items)
mood_labels = list(mood_labels[i] for i in selected_items)

df_final_singledataset = pd.DataFrame(df_singledata[mood_columns], \
                                            columns=['mood_' + i for i in mood_labels])
df_final_singledataset['id'] = 1

# Convert date&time to minutes
df_final_singledataset['timing_all'] = prep.time_to_minutes(df_singledata) + 1
df_final_singledataset['timing'] = df_final_singledataset['timing_all'].copy()
df_final_singledataset['days'] = (pd.to_datetime(df_singledata['date'], format='%d/%m/%y') \
                         - pd.to_datetime(df_singledata['date'].iloc[0], format='%d/%m/%y')).dt.days + 1

# Depression score
df_final_singledataset['dep'] = df_singledata['dep']

nonnan_idx = np.sum(df_final_singledataset[mood_columns].isna(),axis=1)==0
df_final_singledataset = df_final_singledataset[nonnan_idx]

# Save as csv file
if savefiles:
    df_final_singledataset.to_csv(filepath + 'data_prep/singledataset/mooddata.csv', \
                           index = False)

change_point = 127 # from paper
df_final_singledataset['id'][df_final_singledataset['days'] < change_point] = 1
df_final_singledataset['id'][df_final_singledataset['days'] >= change_point] = 2
df_final_singledataset['timing'][df_final_singledataset['id']==2] = \
df_final_singledataset['timing'][df_final_singledataset['id']==2] - \
df_final_singledataset['timing'][df_final_singledataset['id']==2].iloc[0] + 1
df_final_singledataset['originalid'] = 0

# Save as csv file
if savefiles:
    df_final_singledataset.to_csv(filepath + 'data_prep/singledataset/split/mooddata.csv', \
                           index = False)

# Split into baseline, discontinuation and after transition
df_final_singledataset['id'] = np.nan
intervals = [[0,28], [28, 98], [127, 197]]
for j, i in enumerate(intervals):
    idx = (df_final_singledataset['days'] > i[0]) \
                                 & (df_final_singledataset['days'] <= i[1])
    df_final_singledataset['id'][idx] = j + 1
    df_final_singledataset['timing'][idx] = df_final_singledataset['timing'][idx] - \
                                            df_final_singledataset['timing'][idx].iloc[0] + 1
    df_final_singledataset['originalid'] = 0

# save as csv file
if savefiles:
    df_final_singledataset.to_csv(filepath + 'data_prep/singledataset/split3/mooddata.csv', \
                           index = False)
```

```python
# Create input matrix
what_columns = [col for col in df_singledata.columns if 'what' in col]
who_columns = [col for col in df_singledata.columns if 'who' in col]
input_columns = [what_columns, who_columns]

what_labels = ['resting', 'work/studies', 'housekeeping/shopping', 'caring for others', 'medical care',
               'taking care of oneself', 'sports', 'active relaxation', 'passive relaxation', 
               'chatting/texting/facebook etc', 'chilling', 'talking', 'eating/drinking', 'traveling', 'other']
what_scales = [1, 10, 20, 21, 26, 27, 41, 43, 45, 47, 49, 51, 60, 88, 89]
who_labels = ['partner', 'family resident', 'roommates', 'family non-resident', 'family living at other places',
              'friends', 'colleagues', 'acquaintances', 'strangers/others']
who_scales = [10, 17, 19, 27, 29, 30, 40, 49, 50]

input_labels = [{what_scales[i]: what_labels[i] for i in range(len(what_labels))}, \
 {who_scales[i]: who_labels[i] for i in range(len(who_labels))}]

df_input = prep.creating_input_df(df_singledata, [1], input_columns, input_labels)

alone = df_input[[i for i in df_input.columns if i in who_labels]].sum(axis=1) == 0
nothing = df_input[[i for i in df_input.columns if i in what_labels]].sum(axis=1) == 0
df_input.loc[alone, 'alone'] = 1
df_input.loc[nothing, 'nothing'] = 1
df_input = df_input.fillna(0)
# df_input_nocoll, foo = prep.remove_collinear_rows(df_input[df_input.columns[2:]].to_numpy())

if savefiles:
    df_input.to_csv(filepath + 'data_prep/singledataset/inputdata.csv', \
                           index = False)

df_input = df_input[nonnan_idx]
df_input['id'][df_final_singledataset['days'] < change_point] = 1
df_input['id'][df_final_singledataset['days'] >= change_point] = 2

df_input['originalid'] = 0

if savefiles:
    df_input.to_csv(filepath + 'data_prep/singledataset/split/inputdata.csv', \
                           index = False)

df_input['id'] = np.nan
intervals = [[0,28], [28, 98], [127, 197]]
for j, i in enumerate(intervals):
    idx = (df_final_singledataset['days'] > i[0]) \
                                 & (df_final_singledataset['days'] <= i[1])
    df_input['id'][idx] = j + 1
    df_input['originalid'] = 0

if savefiles:
    df_input.to_csv(filepath + 'data_prep/singledataset/split3/inputdata.csv', \
                           index = False)
```
