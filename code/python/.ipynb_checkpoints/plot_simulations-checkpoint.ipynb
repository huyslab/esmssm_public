```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

# Define filepath
filepath = '../../results/modelfits/'

# Define path to save figures
figpath = '../../results/paper_figures/'
savefig = False
```

```python
# Import simulation feature file
mat = scipy.io.loadmat('../matlab/simulation/kf_simulations.mat')
```

```python
# Extract simulation features
dotprod, nrmseA, nrmseZ, outlier = [[] for i in range(4)]
for i in ['days', 'dim', 'noise', 'gap']:
    dotprod.append(mat['kf_simulations'][i][0][0]['dotprod'][0][0][0,:,:])
    nrmseA.append(mat['kf_simulations'][i][0][0]['nrmseA'][0][0])
    nrmseZ.append(mat['kf_simulations'][i][0][0]['nrmseZ'][0][0])
```

```python
# Plot simulation for different settings

def plot_boxplot(data,ax):
    data_plot = pd.melt(pd.DataFrame(data))
    hue_plot_params = {
        'data': data_plot,
        'x': 'variable',
        'y': 'value',
        "showfliers": False,
        "medianprops": dict(color="orange"),
        "color": 'white',
    }
    b = sns.boxplot(**hue_plot_params, ax=ax)

sns.set(font_scale=1.5)
xticklabels = [[2, 6, 10, 14], [2,4,8,16], [0.1, 0.5, 1, 2, 5], ['fast', 'slow']]
xlabel=['days of\nexperience sampling', 'emotion\ndimensionality', 'variance of\nmeasurement noise', 'latent process speed']
fig, axs = plt.subplots(3,4, figsize=(18,15))
for i in range(4):

    # Boxplot NRMSE of known and re-estimated latent states
    plot_boxplot(nrmseZ[i],axs[0,i])
    axs[0,i].set(xticklabels=xticklabels[i],xlabel=None,ylabel=None)
    if i == 0: 
        axs[0,i].set_ylabel('NRMSE of known and\nestimated latent states', fontsize=20)

    # Boxplot NRMSE of known and re-estimated A matrix
    plot_boxplot(np.real(nrmseA[i]),axs[1,i])
    axs[1,i].set(xticklabels=xticklabels[i],xlabel='',ylabel='')
    if i == 0: 
        axs[1,i].set_ylabel('NRMSE of known and\nestimated dynamics matrices', fontsize=20)

    # Boxplot dotproduct of known and re-estimated eigenvectors of A matrix
    plot_boxplot(dotprod[i],axs[2,i])
    axs[2,i].set(xticklabels=xticklabels[i], ylim=(-0.1,1.1),ylabel=None)
    axs[2,i].set_xlabel(xlabel[i], fontsize=20)
    if i == 0: 
        axs[2,i].set_ylabel('absolute dot product of\n known and estimated\nfirst eigenvectors', fontsize=20)

for i in range(3):
    for j in range(4):
        lab = ['A', 'B', 'C', 'D'][j]
        axs[i, j].text(-0.15, 1, lab + str(i+1), transform=axs[i, j].transAxes, size=20, weight='bold')
        
plt.tight_layout()
if savefig: plt.savefig(figpath + 'malamud_fig2_simulation.svg', bbox_inches='tight')
```
